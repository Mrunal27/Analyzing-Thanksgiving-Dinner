```python
import pandas as pd
import numpy as np

data = pd.read_csv('thanksgiving.csv',encoding = "Latin-1")

data.head()

```

```python
columns = data.columns
```

```python
celebrates = data[data['Do you celebrate Thanksgiving?'] == "Yes"]
celebrates_count = celebrates['Do you celebrate Thanksgiving?'].value_counts()
celebrates_count
```

```python
main_dish = data['What is typically the main dish at your Thanksgiving dinner?']
main_dish_dict = dict()

for each in main_dish:
    main_dish = celebrates[celebrates['What is typically the main dish at your Thanksgiving dinner?'] == each]
    main_dish_count = main_dish['What is typically the main dish at your Thanksgiving dinner?'].value_counts()
    main_dish_dict[each] = main_dish_count

main_dish_count
```

```python
have_gravy = data['Do you typically have gravy?']

only_Tofurkey = data[data['What is typically the main dish at your Thanksgiving dinner?'] == 'Tofurkey']

for each in have_gravy:
    gravy_and_Tofurkey = only_Tofurkey[only_Tofurkey['Do you typically have gravy?'] == each]

gravy_and_Tofurkey
```

```python
apple = data['Which type of pie is typically served at your Thanksgiving dinner? Please select all that apply. - Apple']
pumpkin = data['Which type of pie is typically served at your Thanksgiving dinner? Please select all that apply. - Pumpkin']
pecan = data['Which type of pie is typically served at your Thanksgiving dinner? Please select all that apply. - Pecan']

apple_isnull = pd.isnull(apple)
pumpkin_isnull = pd.isnull(pumpkin)
pecan_isnull = pd.isnull(pecan)

no_pies = apple_isnull & pumpkin_isnull & pecan_isnull

print(no_pies.value_counts())

apple_count = len(no_pies[apple_isnull])
pumpkin_count = len(no_pies[pumpkin_isnull])
pecan_count = len(no_pies[pecan_isnull])

apple_count, pumpkin_count, pecan_count
```

```python
def convert_age(strage):
    if(pd.isnull(strage)):
        return None
    agelist = strage.split(" ") 
    first_elm = agelist[0]
    age = int(first_elm.replace("+",""))
    return age


int_age = data['Age'].apply(convert_age)
data['int_age'] = int_age

int_age.describe()
```

```python
def convert_income(strincome):
    if(pd.isnull(strincome)):
        return None
    incomelist = strincome.split(" ")
    firstelm = incomelist[0]
    
    if(firstelm == "Prefer"):
        return None
    
    firstelm = firstelm.replace('$','').replace(',','')
    result = int(firstelm)
    return result


int_income = data['How much total combined money did all members of your HOUSEHOLD earn last year?'].apply(convert_income)
data['int_income'] = int_income
int_income.describe()
```

```python
data[data["int_income"] < 150000]["How far will you travel for Thanksgiving?"].value_counts()
```
